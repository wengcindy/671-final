```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Performance (loss only)

```python
performance_loss = pd.read_csv("deepOBS performance loss.csv")
performance_loss
```

```python
models = list(performance_loss.columns)
models = models[1:]
models
```

```python
max_performance = max(performance_loss.select_dtypes(include=[np.number]).max())
min_performance = min(performance_loss.select_dtypes(include=[np.number]).min())
```

```python
ax = plt.gca()

for m in models:
    performance_loss.plot(kind='line',x='Task',y=m,ax=ax)
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.xticks(rotation=90)
plt.yticks(np.arange(int(min_performance),max_performance, 5))
plt.title('Performance')
plt.xlabel('Tasks')
plt.ylabel('Loss')
plt.show()
```
