```python
crowded_valley = pd.read_csv("Crowded_Valley_results/large_budget/none/best_perf_avg_final_test_accuracy.csv")
crowded_valley.rename(columns={'Unnamed: 0': 'objective'}, inplace=True)
crowded_valley.set_index('objective', inplace=True)
crowded_valley
```

```python
test_losses = np.asarray(crowded_valley)
test_losses
normalized_test_losses = []

for i in range(len(test_losses)):
    mean = np.mean(test_losses[i])
    minus_mean = test_losses[i] - mean
    normalized_test_losses.append((minus_mean)/np.linalg.norm(minus_mean))
print(normalized_test_losses)
```

```python
def minus_mean_euclid_norm(loss_array, axis=1):
    num_rows = loss_array.shape[0]
    mean = np.reshape(np.mean(loss_array, axis=axis), (num_rows, 1))
    minus_mean = loss_array - mean
    norm_array = np.reshape(np.linalg.norm(minus_mean, axis=axis), (num_rows, 1))
    return minus_mean/norm_array
```
