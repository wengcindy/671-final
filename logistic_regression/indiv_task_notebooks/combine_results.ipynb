```python
import pandas as pd
```

```python
df1 = pd.read_csv('logistic_regression_cifar100_normalized_loss.csv')
df2 = pd.read_csv('logistic_regression_cifar10_normalized_loss.csv')
df3 = pd.read_csv('logistic_regression_fmnist_normalized_loss.csv')
df4 = pd.read_csv('logistic_regression_gaussian_normalized_loss.csv')
df5 = pd.read_csv('logistic_regression_mnist_normalized_loss.csv')
df6 = pd.read_csv('logistic_regression_svhn_normalized_loss.csv')
```

```python
frames = [df1, df2, df3, df4, df5, df6]
```

```python
result = pd.concat(frames)
result.rename(columns={'Unnamed: 0': 'objective'}, inplace=True)
result.set_index('objective', inplace=True)
result
```

```python
result.to_csv('task_normalized_losses.csv')
```

```python
df1 = pd.read_csv('logistic_regression_cifar100_loss.csv')
df2 = pd.read_csv('logistic_regression_cifar10_loss.csv')
df3 = pd.read_csv('logistic_regression_fmnist_loss.csv')
df4 = pd.read_csv('logistic_regression_gaussian_loss.csv')
df5 = pd.read_csv('logistic_regression_mnist_loss.csv')
df6 = pd.read_csv('logistic_regression_svhn_loss.csv')
```
